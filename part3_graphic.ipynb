문제1)  평균 구하기

다음과 같은 학생들의 수학 점수가 있다. 평균 점수를 구하시오.

```python
import pandas as pd

data1 = {'이름': ['철수', '영희', '민수', '지영', '수지'],
        '수학': [88, 92, 75, 95, 80]}
df1 = pd.DataFrame(data1)
print(df1)

# 수학 점수의 평균 구하기
average = df1['수학'].mean()
print(average)
```

```python
# 문제2) 중앙값 구하기

# 다음 데이터에서 키의 중앙값을 구하시오.

data2 = {'이름': ['A', 'B', 'C', 'D', 'E', 'F'],
        '키': [170, 165, 180, 175, 160, 185]}
df2 = pd.DataFrame(data2)

# 키 중앙값 구하기
# 중앙값 == 중위수 구하기(짝수이므로, 중간값이 2개라 2개의 평균을 구함)
mid = df2['키'].median()
print(mid)
```

```python
# 문제3) 최빈값 구하기

# 다음 고객 데이터에서 구매한 상품 수의 최빈값을 구하시오.

data3 = {'고객ID': [101, 102, 103, 104, 105, 106, 107,108],
        '구매_상품수': [2, 3, 3, 2, 4, 2, 3, 3]}
df3 = pd.DataFrame(data3)

# 구매한 상품의 최빈값 구하기
# 최빈값 == 원소 빈도가 많은 것 구하기

df3['구매_상품수'].mode()
```

```python
# 문제4) 분산과 표준편차

# 다음 데이터를 사용해서 몸무게의 분산과 표준편차를 각각 구하시오.

data4 = {'이름': ['홍길동', '임꺽정', '장길산', '이순신'],
        '몸무게': [65, 70, 75, 80]}
df4 = pd.DataFrame(data4)

# 몸무게 분산
variance = df4["몸무게"].var()
print('몸무게 분산: ', variance)

# 몸무게 표준편차
standard = df4["몸무게"].std()
standard
```

```python
# 문제5) describe() 함수 활용

# 다음 직원의 근무 시간 데이터가 있다. describe()를 이용해 전체 요약 통계를 출력하고, 그 중 평균과 중앙값을 따로 확인하시오.

data5 = {'직원': ['A', 'B', 'C', 'D', 'E', 'F'],
        '주간_근무시간': [38, 40, 42, 37, 45, 41]}
df5 = pd.DataFrame(data5)
print(df5)
print('-'*100)

# 전체 요약 통계
des5 = df5.describe()
print(des5)
print()

# 평균
average = df5.mean(numeric_only=True)
print(average)
print()

# 중앙값
df5.median(numeric_only=True)
```
