```python
# 전체 상관계수 구하기
import numpy as np

df5_corr = df5.corr()
index_size = len(df5_corr.index) # 행의 개수 16개
col_size = len(df5_corr.columns) # 열의 개수 16개

for i in range(0, index_size):
    for j in range(0, col_size):
        if np.abs( df5_corr.iloc[i,j] ) >= 0.5 :
            print( df5_corr.iloc[i,j] )
        
```

```python
# 상관계수 절대값이 0.5 이상인 것들의 열 이름을 리스트로 저장하기
import numpy as np


df5_corr = df5.corr()
df5_corr
# 상관계수 값을 가져와서 0.5 이상인지 판단
#행수
df5_index_len = len(df5_corr.index) # 행의 개수
df5_index_len

# 열의 개수
df5_columns_len = len(df5_corr.columns)
df5_columns_len

# 열명들을 리스트에 담았다
df5_corr_col_name = df5_corr.columns
print( df5_corr_col_name )

result_list = []

for i in range(df5_index_len): # i 값은 0,1,2,..15
    for j in range(df5_columns_len): # j = 0,1,2,..15
        if( (np.abs( df5_corr.iloc[i][j])) >= 0.5 and (np.abs(df5_corr.iloc[i][j])) != 1 ):
            print( df5_corr.iloc[i][j] )
            result_list.append( df5_corr_col_name[j] )
    print('\n')


# 중복 제거를 위해 set() 넣기
result_list = set(result_list)
# set 자료를 list로 만들기
result_list = list(result_list)
result_list
    
# import numpy
# corr_abs = numpy.abs(corr_values) # 절대값으로 저장
# for i in corr_abs:
#     for ii in i:
#         if ii >= 0.5 :
#             df.iloc[ ]
```
