```python
x = [1,2,3,4]
y = [2,3,5,10]
plt.plot(x,y)
plt.fill_between(x[2:4], y[2:4], alpha=0.5)
        # x[:-1]
plt.show()                
```

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1, 'm-')
plt.plot(x,y2)
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='pink', alpha=0.5)
        # x[:-1]
# x[1:3]   a1=2, a2=3
# y1[1:3]   b1=3, b2=5
# 1 -     (a1,b1) (a2,b2)  (2,3), (3,5)
# y2[1:3]   1=2, 2=4
# 2 -     (a1,c1), (a2,b2) -  - (a2,c2) 
plt.show()                
```

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='pink', alpha=0.5)
#  (1.9, 1.0), (1.9, 4.0), (3.1, 6.0), (3.1, 3.0)    
plt.show()                
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
print(x1)
x2 = np.linspace(0.0, 5.0)
print(x2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) # np.pi (3.14) / np.cos  / np.exp  
print( y1)
print("-"*100)

y2 = np.cos(2 * np.pi * x2)
print(y2)


fig, axes = plt.subplots(  1,2, figsize=(12,5) )
axes[0].plot( x1, y1 )
axes[1].plot( x2, y2 )
plt.show()

```
