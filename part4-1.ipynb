# 시각화 도구

```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_excel('./data-4/시도별_전출입_인구수.xlsx')
df.tail()
```

```python
df = df.ffill()
df.tail()
```

```python
mask =(df['전출지별'] == '서울특별시') & ( df['전입지별'] != '서울특별시' )
df_seoul = df[mask]
df_seoul

# 전출지별 열 삭제하기
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul

# 전출지별 --> 전출지 이름 변경
df_seoul = df_seoul.rename( {'전입지별' : '전입지'}, axis=1 )
df_seoul

# 전입지 열을 인덱스로 변경하기
df_seoul = df_seoul.set_index('전입지')
df_seoul

# 각 열에 대한 원소 형태를 보기
df_seoul.info()
df_seoul

# 
```

```python
# 서울 -> 경기도
import pandas as pd
import matplotlib.pyplot as plt
sr_one = df_seoul.loc['경기도']
print(sr_one.head())
sr_one.index
sr_one.values
sr_one.plot(x= sr_one.index, y = sr_one.values, kind='line')
```

```python
plt.plot( sr_one.index, sr_one.values, linestyle=':')
```

```python
plt.plot(sr_one.index, sr_one.values, linestyle="--")
plt.title('서울-> 경기 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show()
```

# KOSIS 다운 인구통계 자료로 [제주 -> 서울 인구 이동] 구하기

```python
# matplotlib 한글 폰트 오류 해결
import pandas as pd
from matplotlib import pyplot as plt


from matplotlib import font_manager, rc
font_path = './data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_excel('./data-4/clean전출지_전입지_시도_별_이동자수_20250428160441.xlsx')
df
```

```python
df = df.ffill()
df
```

```python
mask = ( df['전출지별'] == '제주특별자치도' ) & ( df['전입지별'] != '제주특별자치도')
df_jeju = df[mask]
print(df_jeju)
print('-'*100)

df_jeju = df_jeju.drop(['전출지별'], axis=1)
print(df_jeju)
print('-'*100)

df_jeju = df_jeju.rename({'전입지별':'전입지'}, axis=1)
print(df_jeju)

df_jeju = df_jeju.set_index('전입지')
print('-'*100)
```

```python
sr_two = df_jeju.loc['서울특별시']
print(sr_two)
print('-'*100)

print(plt.plot(sr_two.index, sr_two.values))
print('-'*100)



print(plt.title('< 제주 -> 서울 인구 이동 >'))
print('-'*100)

plt.xlabel('기간')
plt.ylabel('이동자수')
plt.legend(labels=['제주 -> 서울'], loc='best')



plt.show()
```

```python
plt.style.available
```

```python


plt.figure(figsize=(14,5))

plt.xticks(size=15, rotation=0)

print(plt.plot(sr_two.index, sr_two.values))
print('-'*100)



plt.plot(sr_two.index, sr_two.values,
        marker='o',
        markerfacecolor='mediumturquoise',
        markeredgecolor='pink',
        markeredgewidth=2,
         markersize=10,
         linestyle='-',
         color='pink',
         label='제주 -> 서울'
        )



print(plt.title('< 제주 -> 서울 인구 이동 >', size=30))
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

plt.legend(loc='best', fontsize=15)

plt.show()
```
