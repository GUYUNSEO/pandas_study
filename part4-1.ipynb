# KOSIS 다운 인구통계 자료로 [제주 -> 서울 인구 이동] 구하기

```python
# matplotlib 한글 폰트 오류 해결
import pandas as pd
from matplotlib import pyplot as plt


from matplotlib import font_manager, rc
font_path = './data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_excel('./data-4/clean전출지_전입지_시도_별_이동자수_20250428160441.xlsx')
df
```

```python
df = df.ffill()
df
```

```python
mask = ( df['전출지별'] == '제주특별자치도' ) & ( df['전입지별'] != '제주특별자치도')
df_jeju = df[mask]
print(df_jeju)
print('-'*100)

df_jeju = df_jeju.drop(['전출지별'], axis=1)
print(df_jeju)
print('-'*100)

df_jeju = df_jeju.rename({'전입지별':'전입지'}, axis=1)
print(df_jeju)

df_jeju = df_jeju.set_index('전입지')
print('-'*100)
```

```python
sr_two = df_jeju.loc['서울특별시']
print(sr_two)
print('-'*100)

print(plt.plot(sr_two.index, sr_two.values))
print('-'*100)



print(plt.title('< 제주 -> 서울 인구 이동 >'))
print('-'*100)

plt.xlabel('기간')
plt.ylabel('이동자수')
plt.legend(labels=['제주 -> 서울'], loc='best')



plt.show()
```

```python
plt.style.available
```

```python


plt.figure(figsize=(14,5))

plt.xticks(size=15, rotation=0)

print(plt.plot(sr_two.index, sr_two.values))
print('-'*100)



plt.plot(sr_two.index, sr_two.values,
        marker='o',
        markerfacecolor='mediumturquoise',
        markeredgecolor='pink',
        markeredgewidth=2,
         markersize=10,
         linestyle='-',
         color='pink',
         label='제주 -> 서울'
        )



print(plt.title('< 제주 -> 서울 인구 이동 >', size=30))
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
# 195p
# 서울에서 충청남도 경상북도 강원도 전라남도 이동한 인구 데이터
a = range(1970, 2018) # 1970, 1971, ... , 2017년 반복해서 만들어 내는 반복자
print(a)

print(df_seoul.tail())
print('ㅡ'*80)

col_years = list( map(str, range(1970,2018) ) )
col_years

df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_4
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,20)) # 캠퍼스 사이즈

ax1 = fig.add_subplot(3,2,1)
ax2 = fig.add_subplot(3,2,2)
ax3 = fig.add_subplot(3,2,3)
ax4 = fig.add_subplot(3,2,4)
ax5 = fig.add_subplot(3,2,5)


ax1.plot( col_years, df_4.loc['충청남도', :], marker='o', markerfacecolor='green', markersize=10, color='olive', linewidth=2, label='서울 -> 충남' )
ax2.plot(col_years, df_4.loc['경상북도', :], marker='o', markerfacecolor='blue', markersize=10, color='skyblue', linewidth=2, label='서울 -> 경북')
ax3.plot(col_years, df_4.loc['강원도', :], marker='o', markerfacecolor='red', markersize=10, color='magenta', linewidth=2, label='서울 -> 강원')
ax4.plot(col_years, df_4.loc['전라남도', :], marker='o', markerfacecolor='orange', markersize=10, color='yellow', linewidth=2, label='서울 -> 전남')

         
ax5.plot(col_years, df_4.loc['전라남도', :], marker='o', markerfacecolor='orange', markersize=10, color='yellow', linewidth=2, label='서울 -> 전남')
ax5.plot(col_years, df_4.loc['충청남도', :], marker='o', markerfacecolor='green', markersize=10, color='yellow', linewidth=2, label='서울 -> 전남')
ax5.plot(col_years, df_4.loc['경상북도', :], marker='o', markerfacecolor='blue', markersize=10, color='yellow', linewidth=2, label='서울 -> 전남')
ax5.plot(col_years, df_4.loc['강원도', :], marker='o', markerfacecolor='red', markersize=10, color='yellow', linewidth=2, label='서울 -> 전남')


ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
ax5.legend()


ax1.set_title('< 서울 -> 충남 인구이동 >', size=15)
ax2.set_title('< 서울 -> 경북 인구이동 >', size=15)
ax3.set_title('< 서울 -> 강원 인구이동 >', size=15)
ax4.set_title('< 서울 -> 전남 인구이동 >', size=15)
ax5.set_title('< 서울 -> 충남, 경북, 강원, 전남 인구이동 >', size=15)



ax1.set_xticks( range(len(col_years)) ) # 0,1,2,...,47 x축에 47
ax1.set_xticklabels( col_years, rotation=90 )

ax2.set_xticks( range(len(col_years)) )
ax2.set_xticklabels( col_years, rotation=45, fontsize=7 )

ax3.set_xticks( range(len(col_years)) )
ax3.set_xticklabels( col_years, rotation=45, fontsize=7 )

ax4.set_xticks( range(len(col_years)) )
ax4.set_xticklabels( col_years, rotation=45, fontsize=7 )

ax5.set_xticks( range(len(col_years)) )
ax5.set_xticklabels( col_years, rotation='vertical', fontsize=7 )


ax1.set_ylim(10000, 60000)
ax2.set_ylim(10000, 60000)
ax3.set_ylim(10000, 60000)
ax4.set_ylim(10000, 60000)
ax5.set_ylim(10000, 60000)

plt.show()
```
